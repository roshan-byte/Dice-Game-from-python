```python
#                                        Dice Game   @author roshan-byte

from IPython.display import clear_output
import random

def rollDice(choice):
    if(choice >=7):
        print("\n\nwrong guess")
        print("\nplease guess number between 1 to 6 and try again !!!!!!!!!")
        return
    for i in range(1,15):
        print("-",end="  ")
    print()
    
    a=random.randint(1,6)
    
    if(a==1):
        for i in range(1,15):
                if(i==7):
                    print("|               *                  |")
                else:
                    print("|                                  |")
    elif(a==2):
        for i in range(1,15):
            if(i==6 or i==9):
                print("|                   *                   |")
            else:
                print("|                                       |")
    
    elif(a==3):
        for i in range(1,15):
            if(i==6 or i==8 or i==10):
                print("|                   *                   |")
            else:
                print("|                                       |")
                
    elif(a==4):
        for i in range(1,15):
            if(i==7 or i==10):
                print("|            *             *            |")
            else:
                print("|                                       |")
                
    elif(a==5):
        for i in range(1,15):
            if(i==5):
                print("|            *             *            |")
            elif(i==8):
                print("|                   *                   |")
            elif(i==11):
                print("|            *             *            |")
            else:
                print("|                                       |")
                
    elif(a==6):
        for i in range(1,15):
            if(i==5 or i==8 or i==11):
                print("|            *             *            |")
            else:
                print("|                                       |")

    for i in range(1,15):
            
        print("-",end="  ")
        
    print()
    
    
    if(a==choice):
        print("\n\n congratulations you win the game\r")
        
        clear_output(wait=True)
    else:
        print("\n\nAhhhh you lose the game but don't lose hope try again\r")
        clear_output(wait=True)


        
print("                              WELCOME TO THE ROLLING GAME OF DICE")

while(True):
    choice=input("Enter 'yes' to continue or enter 'no' to end the game  ")
    if(choice=="yes"):
        number=int(input("\nselect any number between 1 to 6 and roll the dise \n"))
        rollDice(number)
    else:
        break

```
